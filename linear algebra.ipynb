```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
```

# Functions

```python
def task(imgUrl):
    img = mpimg.imread(imgUrl)
    fig,ax=plt.subplots(1,1,figsize=(12,4))
    ax.imshow(img)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```

# Vectors

```python
# Task 1
task(r'linear algebra exercises\vectors\exer1.PNG')
```

```python
# a 
a1=np.array([4,5,1,0])
a2=np.array([-4,-3,3,10])
a=a1+a2

# b
b1=np.array([4,2,0])
b2=np.array([6,-4,60])
b3=np.array([2,-5,40])
b=b1-b2+b3

# c
c1=np.array([1,0])
c2=np.array([1,2])
c=c1+c2

# d
d1=np.array([2,2])
d2=np.array([3,4])
d=d1-d2

# e
e1=np.array([-3,1])
e2=np.array([3,-1])
e=e1+e2

# f
f1=np.array([1,4])
f2=np.array([2,8])
f=f1+f2

# print result
print(pd.DataFrame({'Vectors':['a','b','c','d','e','f'],'Result':[a,b,c,d,e,f]}))

# plot 2D vectors
plt.plot([0,2],[0,2],'r.-',label='vector c')
plt.plot([0,-1],[0,-2],'b.-',label='vector d')
plt.plot([0,0],[0,0],'k.-',label='vector e')
plt.plot([0,3],[0,12],'g.-',label='vector f')
plt.axis('square')
plt.xlim([-7,7])
plt.legend()
plt.show()
```
